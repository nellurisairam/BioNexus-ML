```python
import streamlit_authenticator as stauth

passwords = ['1234', '1111']

# 1. Initialize the Hasher object without arguments
hasher = stauth.Hasher(passwords) 

# 2. Call generate (some versions may require passwords here instead)
hashed_pw = hasher.generate()

print(hashed_pw)
```
